# Subset mutations.tsv to top 25 mutations (Table 2)

```python
import pandas as pd
import numpy as np
```

```python
mutations = pd.read_csv("paper/mutations.tsv", sep="\t", index_col=0)
mutations = mutations[mutations.index <= 25]
mutations = mutations[['mutation', 'mean/stddev', 'Δ log R', 'emerged in lineages']]

mutations['Δ log R'] = [int(1000 * x) / 1000.0 for x in mutations['Δ log R']]
mutations['emerged in lineages'] = np.array([len(s.split(',')) for s in mutations['emerged in lineages']])
mutations['Gene'] = [s.split(':')[0] for s in mutations['mutation']]
mutations['Mutation'] = [s.split(':')[1] for s in mutations['mutation']]
mutations = mutations[['Gene', 'Mutation', 'Δ log R', 'emerged in lineages']]
mutations.columns = ['Gene', 'Mutation', 'Fold increase in growth rate', 'Number of Lineages']
```

```python
mutations.to_csv('mutations_subset.tsv', sep='\t')
```
